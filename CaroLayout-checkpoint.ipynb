```python
import pygame, itertools
import sys
from pygame.locals import QUIT, MOUSEBUTTONUP
import random, os
```

```python
#khai báo các màu có thể dùng trong bàn cờ
GRAY=(128,128,128)
NAVYBLUE=(60,60,100)
WHITE=(255,255,255)
RED=(255,0,0)
GREEN=(0,255,0)
BLUE=(0,0,255)
YELLOW=(255,255,0)
ORANGE=(255,128,0)
PURPLE=(255,0,255)
CYAN=(0,255,255)
CYAN2=(54, 199, 205, 0.8)
GREEN2=(125, 169, 172, 0.51)
```

```python
class Square: #lớp ô vuông , đối tượng ô vuông thuộc bàn cờ
    state=0 #trạng thái khi chưa được đánh dấu
    def __init__(self,x,y,size,board):
        self.size=size
        self.line_width=int(self.size / 40) if self.size > 40 else 1
        self.radius=self.size*1/4
        self.radius1=self.size*1/3
        self.rect=pygame.Rect(x,y,size,size)
        self.board=board
    def mark_x(self):#đánh xấu x
            pygame.draw.line(self.board.surface,ORANGE,(self.rect.centerx-self.radius,self.rect.centery-self.radius),(self.rect.centerx+self.radius,self.rect.centery+self.radius),self.line_width)
            pygame.draw.line(self.board.surface,ORANGE,(self.rect.centerx+self.radius,self.rect.centery-self.radius),(self.rect.centerx-self.radius,self.rect.centery+self.radius),self.line_width)
    def mark_o(self):#đánh dấu o
            pygame.draw.circle(self.board.surface,GRAY,(self.rect.centerx,self.rect.centery),self.radius1,self.line_width)
```

```python
class Board(object): #lớp bàn cờ
    turn=1 #X đi trước
    def __init__(self,grid_size=3,box_size=25,border=1,line_width=1):#constructor
        self.grid_size=grid_size
        self.box_size=box_size
        self.border=border
        self.line_width=line_width
        surface_size = (self.grid_size * self.box_size) + (self.border * 2) + (self.line_width * (self.grid_size - 1))
        self.surface = pygame.display.set_mode((surface_size, surface_size), 0, 256)
        self.end=False
        self.setup()
    def setup(self):#cài đặt bàn cờ
        pygame.display.set_caption("game caro")
        icon=pygame.image.load('caroicon.png')
        pygame.display.set_icon(icon)
        self.surface.fill(WHITE)
        self.draw_lines()
        self.create_squares()
    def draw_lines(self):#vẽ các đường kẻ phân chia các ô
        for i in range (1,self.grid_size):
            start_position = ((self.box_size * i) + (self.line_width * (i - 1))) + self.border
            width = self.surface.get_width() - (2* self.border)
            pygame.draw.rect(self.surface, CYAN2, (start_position, self.border, self.line_width, width))
            pygame.draw.rect(self.surface, CYAN2, (self.border, start_position, width, self.line_width))
    def create_squares(self):#tạo ra các đối tượng ô vuông giữa những đường kẻ
        self.list_square=[] #khởi tạo danh sách tất cả các đối tượng ô vuông trên bàn cờ
        self.list_x=[] #danh sách dấu x đã đánh
        self.list_o=[] #danh sách dấu o đã đánh

        list_start_square=[] #danh sách các vị trí bắt đầu (góc trên bên trái) theo 1 chiều
        for i in range(0,self.grid_size):
            num=((i*self.box_size)+self.border+(i*self.line_width))
            list_start_square.append(num)
        all_square=list(itertools.product(list_start_square,repeat=2)) #hàm đưa ra tất cả cặp x,y từ 1 danh sách
        for x,y in all_square:
            self.list_square.append(Square(y,x,self.box_size,self)) #tạo đối tượng ô vuông và lần lượt thêm vào danh sách
    def handleClick(self,x,y):
        for index,square in enumerate(self.list_square):
            if square.rect.collidepoint(x,y):
                self.player_turn(index,square)
    #lượt người chơi
    def player_turn(self,index,square): #truyền vào 3 tham số: bàn cờ, chỉ số của ô đang click, đối tượng ô vuông đang click
        if square.state!=0:#nếu ô từng được chọn thì bỏ qua
            return
        if self.turn ==1:#nếu là lượt của người thứ 1 thì mark_x
            square.mark_x()
            square.state=1
            self.turn=2
            self.list_x.append(index)
            w=check_win(self,index,1)
            if w==1:
                self.display_game_over(1)
        elif self.turn==2:#nếu là lượt của người thứ 2 thì mark_y
            square.mark_o()
            square.state=2
            self.turn=1
            self.list_o.append(index)
            w=check_win(self,index,2)
            if w==2:
                self.display_game_over(2)
        return  #bổ qua những trường hợp khác
    def display_game_over(self, winner):
        surface_size = self.surface.get_height()
        font = pygame.font.Font('freesansbold.ttf', int(surface_size / 8))
        if winner:
            text = 'Player %s won!' % winner
        else:
            text = 'Draw!'
        text = font.render(text, True, ORANGE, WHITE)
        rect = text.get_rect()
        rect.center = (surface_size / 2, 200)
        self.surface.blit(text, rect)
        self.end=True
```
